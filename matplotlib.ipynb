```python
from  matplotlib import pyplot as plt
import pandas as pd
```

* Get data currency btc  from file csv and mad first colum is index then transition dtype index to datetime 

```python
df = pd.read_csv(r"path\BTCUSDT.csv",index_col=0,)
df.set_index(pd.to_datetime(df.index),inplace=True)
df
```

# plot Volume price  from (2024-01-01) to (2024-03-18) .
* I mad changes figure size to (14,5) , line volume red and add marker "." to slope angle add line kind of grid to chart by function grid ,and style appearance grid lines : label color  red  by parameter  abelcolor  and leble size medium and sold = 3 by parameter width

```python
fig, ax = plt.subplots(figsize=(14,5))


ax.plot(df[df.index > pd.to_datetime("2024-01-01")].Volume, color='red', marker='.', linestyle='solid')

ax.set(xlabel='date', ylabel='Volume',
       title='plot Volume price  from (2024-01-01) to (2024-03-18) ')
ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='r', labelsize='medium', width=3)
```
